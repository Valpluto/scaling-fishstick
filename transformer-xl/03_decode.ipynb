```python
import torch
from pytorch_transformers import TransfoXLTokenizer, TransfoXLModel, TransfoXLLMHeadModel
import subprocess

import logging
logging.basicConfig(level=logging.INFO)
```

```python
proc = subprocess.Popen(["head -1 ../data/transformer-xl/low_resource/test-input-text.txt"], stdout=subprocess.PIPE, shell=True)
(out, err) = proc.communicate()
print(out)
```

```python
tokenizer = TransfoXLTokenizer.from_pretrained('transfo-xl-wt103')
```

```python
out="I'm now a Professor of Applied Quantitative Analysis in the Analysis, Engineering, Simulation and Optimization of Performance (AESOP) group in the Department of Computing at Imperial. My broad area of research interest is the application of mathematical modelling techniques to real life systems. Specific areas of interest include, but are not limited to, modelling and optimisation in parallel queueing systems (especially split-merge and fork-join systems), modelling of storage systems, stochastic modelling of sport, stochastic modelling of healthcare systems, resource allocation and control in cloud-computing environments, numerical solution of (semi-)Markov models and specification techniques for SLA specification, compliance prediction and monitoring."

tokenized_text_1 = tokenizer.tokenize(str(out))

# Convert token to vocabulary indices
indexed_tokens_1 = tokenizer.convert_tokens_to_ids(tokenized_text_1)

# Convert inputs to PyTorch tensors
tokens_tensor_1 = torch.tensor([indexed_tokens_1])
```

```python
model = TransfoXLLMHeadModel.from_pretrained('transfo-xl-wt103')
model.eval()
```

```python
max_predictions = 50
mems = None
for i in range(max_predictions):
    predictions, mems = model(tokens_tensor_1, mems=mems)
    predicted_index_tensor = torch.topk(predictions[0, -1, :],5)[1][1] 
    predicted_index = predicted_index_tensor.item()
    predicted_token = tokenizer.convert_ids_to_tokens([predicted_index])[0]
    print(predicted_token)
    
    tokens_tensor_1 = torch.cat((tokens_tensor_1, predicted_index_tensor.reshape(1, 1)), dim=1)
```
