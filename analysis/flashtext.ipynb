```python
import pandas as pd
from flashtext import KeywordProcessor
```

```python
discharge_typos = pd.read_csv('data/discharge_typos.txt', sep='\t',header=None)
discharge_typos.head(20)
```

```python
my_dict = {row[0]: row[1] for row in discharge_typos.values}
```

```python
keywords = list(my_dict.keys())
```

```python
clean = list(my_dict.values())
```

```python
clean
```

```python
processor = KeywordProcessor()

for keyword_name, clean_name in zip(keywords, clean):
    processor.add_keyword(keyword_name, clean_name)
```

```python
document = """blah blah ehco fibroma tropinins fibromama blah """
```

```python
found = processor.replace_keywords(document)
found
```
